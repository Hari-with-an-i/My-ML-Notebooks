```python
# Methods to fill Null values

games_1 = games.copy()
games_2 = games.copy()
games_3 = games.copy()
games_4 = games.copy()
games_5 = games.copy()
games_6 = games.copy()

games_1.dropna(inplace=True)
games_1.isnull().sum()
games_1.shape
```

```python
games_3.fillna(games_3.mean(), inplace=True)
games_3.isnull().sum()
```

```python
games_4.fillna(games_4.median(), inplace=True)
games_4.isnull().sum()
```

```python
games_5.fillna(method="ffill", inplace=True)
games_5.isnull().sum()
games_5.shape
```

```python
games_6.fillna(method="bfill", inplace=True)
games_6.isnull().sum()
```

##Generally used Pandas inbuilt function set.

**DataFrame and Series Creation:**

1. `pd.DataFrame()`: Create a DataFrame.
2. `pd.Series()`: Create a Series.
3. `df.head()`: Return the first n rows of a DataFrame.
4. `df.tail()`: Return the last n rows of a DataFrame.
5. `df.shape`: Get the dimensions of the DataFrame.
6. `df.info()`: Get information about the DataFrame.
7. `df.describe()`: Generate summary statistics.

**Indexing and Selection:**

8. `df[]`: Select columns by label.
9. `df.loc[]`: Select rows by label.
10. `df.iloc[]`: Select rows by integer position.
11. `df.at[]`: Access a single value using a label.
12. `df.iat[]`: Access a single value using an integer position.

**Filtering and Sorting:**

13. `df[df['column'] > value]`: Filter rows based on a condition.
14. `df.sort_values()`: Sort the DataFrame.
15. `df.sort_index()`: Sort by index.
16. `df.groupby()`: Group data by a column.
17. `df.nlargest(n, 'column')`: Get the n largest values in a column.
18. `df.nsmallest(n, 'column')`: Get the n smallest values in a column.

**Data Cleaning:**

19. `df.drop()`: Remove columns or rows.
20. `df.fillna()`: Fill missing values.
21. `df.dropna()`: Remove rows with missing values.
22. `df.duplicated()`: Find duplicate rows.
23. `df.drop_duplicates()`: Remove duplicate rows.
24. `df.replace()`: Replace values.

**Aggregation and Summary:**

25. `df.mean()`: Compute the mean.

27. `df.sum()`: Compute the sum.
28. `df.min()`: Find the minimum value.
29. `df.max()`: Find the maximum value.
30. `df.std()`: Compute the standard deviation.
31. `df.var()`: Compute the variance.
32. `df.corr()`: Compute correlation.

**Data Transformation:**

36. `df.melt()`: Unpivot a DataFrame.
